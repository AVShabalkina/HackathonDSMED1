## Признак 'Ед. изм. кол. показателя'

```python
# Напишем функцию, которая заменяет различные варианты написания 'фл'.
def new_units(value: object) -> str:
    """Функция для преобразования похожих значений в столбце 'Ед. изм. кол. показателя'
    
    Args: 
        значения типа object

    Returns:
       значения типа str
    """
    
    if value in ['fL', 'FL']:
        return 'фл'
    else:
        return value
```

```python
# Применим эту функцию.
project_df['Ед. изм. кол. показателя'] = project_df['Ед. изм. кол. показателя'].apply(new_units)
```

```python
# Посмотрим, что получилось после обработки.
for i in project_df['Кол. лаб. показатель'].unique():
    print(i, ':', project_df[project_df['Кол. лаб. показатель'] == i]['Ед. изм. кол. показателя'].unique())
```

## Признак 'Дата рождения пациента'

```python
# Функция для высчитывания возраста.
def age(date_of_born: object) -> int: 
    """Функция для извлечения возраста пациента из столбца 'Дата рождения пациента'
    
    Args: 
        значения типа object

    Returns:
       значения типа int
    """
    
    today = date.today() 
    return today.year - date_of_born.year - ((today.month, today.day) < (date_of_born.month, date_of_born.day))
```

```python
# Преобразование признака в формат даты.
project_df['Дата рождения пациента'] = pd.to_datetime(project_df['Дата рождения пациента'])

# Применим функцию.
project_df['Возраст'] = project_df['Дата рождения пациента'].apply(age)

# Удалим столбец 'Дата рождения пациента'.
project_df = project_df.drop(labels='Дата рождения пациента', axis=1)
```

С одной стороны мы можем попробовать выбрать таргетной колонкой колонку `'Диагноз'`, тогда мы будем заниматься задачей предсказания лейкоза. Однако, в наших данных нет здоровых людей, а значит мы не можем обучить модель отличать здоровых от больных. При этом, научить модель определять отдельные типы лейкоза также будет достаточно проблематично, так как больше всего в данных у нас представлен диагноз "Другой уточненный лейкоз", который не является окончательным верифецированным диагнозом. 
Однако все же можно попытаться насинтезировать данные анализов крови для "здоровых" людей и построить модель бинарной классификации: здоров - есть подозрение на лейкоз.

Конечно, в идеале такой результат проставляется не методами машинного обучения, а путем простого сравнения значения показателя с его нормой. Однако данный таргет является наиболее приближенным к тому, чего хочет заказчик.

1) Признак `'ID пациента'` не несёт никакой информации и будет лучше его удалить.

2) Признак `'Диагноз'` будет вносить шум. Т.к. 100% пациентов в таблице с тем или иным диагнозом лейкоза (здоровых нет), несмотря на результаты анализа, т.е. этот диагноз закреплён на ID пациента и остаётся за ним, даже если по анализам у пациента ремиссия, для нас он неинформативен. 4) Обновить информацию в признаке `'Результат'`, т.к. в таблице встречались ошибки, путём сравнения `'Значение'` c 2 новыми признаками (`'Верхняя граница`' и `'Нижняя граница'`) и получить 2 значения в данном признаке: 'Норма' = 0, 'Отклонение' = 1.

### Разделение выборки

### Настройка гиперпараметров моделей

Исследуем четыре основные модели: логистическую регрессию, дерево решений, случайный лес и CatBoost

```python
# Настройка гиперпараметров моделей
model_params = {
    'Logistic_Regression': {
        'model': LogisticRegression(random_state=42),
        'params': {
            'penalty': ['l1', 'l2', 'elasticnet', 'none'],
            'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
            'multi_class': ['auto', 'ovr', 'multinomial']
        }
    },
    
    'Decision_Tree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [4, 5, 6, 7, 8, 9, 10], 
            'min_samples_split': [3, 4, 5, 10] 
        }
    },
    
    'Random_Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'max_features': ['sqrt', 'log2'],
            'max_depth': [4, 5, 6, 7, 8],
            'criterion': ['gini', 'entropy']
        }
    },
    
    'Cat_Boost': {
        'model': CatBoostClassifier(random_state=42, logging_level = 'Silent'),
        'params': {
            'iterations': [100, 200, 300],
            'depth': [6, 8, 10],
            'learning_rate': [0.1, 0.05, 0.01]            
        }
    }
}
```

```python
# Расчитаем лучшие гиперпараметры всех моделей
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=3, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
# Сохраняем результаты
best_results = pd.DataFrame(scores, columns=['model','best_score','best_params'])
best_results    
```

### Построение и сравнение confusion matrix

```python
def confusion_matrix_heatmap(y_test, pred, axes):
    """
    Функция confusion_matrix_heatmap() строит heatmap для confusion_matrix по аргументам:
    y_test - тестовые истинные значения
    pred - предсказанные значения модели
    axes - оси, на которых нужно отрисовать хитмеп
    """
    confus_matrix = confusion_matrix(y_test, pred)
    sns.heatmap(confus_matrix, annot=True, ax=axes, fmt='', cmap='Blues')
    axes.set_xlabel('Предсказанное значение')
    axes.set_ylabel('Истинное значение');
```

## Реализация "срочности" похода к врачу

```python
# Функция комментирующая полученное предсказание
def comment(probability: int) -> str:
    """Функция для определения срочности обращения к врачу на основании предсказания модели CatBoost
    
    Args: 
        вероятность принадлежности объекта к классу 1 (отклонение от нормы)

    Returns:
       срочность необходимости обращения к врачу
    """
    
    if probability > 0.75:
        return "Срочно обратитесь к врачу, который назначил вам этот анализ!"
    elif (probability >= 0.5) and (probability <= 0.75):
        return "Рекомендуем посетить врача, который назначил вам этот анализ, в ближайшее время."
    else:
        return "Результат в пределах референсной нормы. Посетите врача с целью получения заключения."
```

```python
# Получим предсказания нашей машины
predict = cb_model.predict_proba(X_test)

# Выбираем вероятность принадледности к классу 1 (отклонение)
urgency = predict[:, 1]

# Добавим комментарий к результатам тестовой выборки.
x_test_urgency = X_test.copy()
x_test_urgency['Результат'] = (urgency >= 0.5).astype('int')
x_test_urgency['Комментарий'] = urgency
x_test_urgency['Комментарий'] = x_test_urgency['Комментарий'].apply(comment)
```

## Парсинг данных

По общему анализу крови в медицине лейкоз не предсказывается, для этого нужны совершенно другие анализы (например, биопсия), однако мы все же попробуем выявить какую-нибудь закономерность между ОАК и лейкозом. Чтобы реализовать идею, нам необходимо анализировать разом всю совокупность показателей, входящих в разовый анализ крови.
Для этого, прежде всего, мы должны распарсить данные по целому анализу, т.е. преобразовать наш датасет из длинной формы в широкую, где в качестве столбцов будут отдельные показатели из списка ОАК. Так как у нас нет пометок о времени проведения анализа, мы не можем сделать это идеально, однако учитывая, что показатели из одного анализа, скорее всего, вносятся разово друг за другом, мы возьмем эту закономерность за основу при парсинге данных.

```python
data_for_diagnosis.info()
```

```python
# Избавляемся от ненужных столбцов
data_for_diagnosis = data_for_diagnosis.drop(['Единицы измерения', 'Референсный интервал', 'Результат'], axis=1)
data_for_diagnosis.head()
```

```python
# Прописываем функцию, которая будет присваивать нашим показателям номер анализа, из которого он был взят
def create_analysis_column(df):
    current_id = None
    current_values = []
    analysis_number = 1
    analysis_index = []

    for index, row in df.iterrows():
        if row['ID пациента'] != current_id:
            analysis_number += 1
            current_values = []

        if row['Лабораторный показатель'] in current_values:
            analysis_number += 1
            current_values = []

        analysis_index.append(analysis_number)
        current_id = row['ID пациента']
        current_values.append(row['Лабораторный показатель'])

    df.loc[:, 'Analysis_Index'] = analysis_index
    return df
```

```python
# Сортировка датафрейма по столбцу 'ID пациента' с учетом индекса
data_for_diagnosis = data_for_diagnosis.sort_values(by='ID пациента').sort_index()

# Удаление дубликатов
data_for_diagnosis = data_for_diagnosis.loc[~data_for_diagnosis.duplicated(keep='first') | ~data_for_diagnosis.shift().duplicated(keep='first')]

# Применяем функцию по присваиванию номера анализа к датафрейму
data_for_diagnosis = create_analysis_column(data_for_diagnosis)

data_for_diagnosis.head()
```

```python
# Вытягиваем значения количественных показателей в вектор признаков
pivot_df = data_for_diagnosis.pivot(index='Analysis_Index', columns='Лабораторный показатель', values='Значение')

# Сброс индекса для того, чтобы 'Analysis_Index' остался отдельным столбцом
pivot_df.reset_index(inplace=True)

pivot_df.head()
```

## Синтез показателей, характерных для нормы

Так как в представленном датасете все пациенту уже больны лейкозом, нам нужны пациенты, которые здоровы. Мы искусственно насинтезируем таких пациентов, отталкиваясь от "нормальных" показателей крови

Данные распределения значений для показателей клинического анализа крови взяли из крупного исследования турецкой популяции https://www.jceionline.org/download/the-largest-reference-range-study-for-hematological-parameters-from-turkey-a-case-control-study-3625.pdf

### Определение параметров для здоровых лиц

```python
# Определение параметров нормальных показателей в мужской популяции
HGB_mean_M =  152.7
HGB_std_M = 9.2

RBC_mean_M = 5.25
RBC_std_M = 0.37

HCT_mean_M = 44.92
HCT_std_M = 2.49

WBC_mean_M = 6.78
WBC_std_M = 1.53

LY_mean_M = 33.79
LY_std_M = 6.95

PLT_mean_M = 238.5
PLT_std_M = 46.95

# Определение параметров нормальных показателей в женской популяции
HGB_mean_F =  128.8
HGB_std_F = 9.2

RBC_mean_F = 4.52
RBC_std_F = 0.36

HCT_mean_F = 39.01
HCT_std_F = 2.51

WBC_mean_F = 6.88
WBC_std_F = 1.78

LY_mean_F = 31.2
LY_std_F = 7.93

PLT_mean_F = 255.9
PLT_std_F = 55.33
```

```python
# Заполнение пропусков датасета с лейкозами значениями нормы
df_cleaned.loc[df_cleaned['Пол']==1, 'HGB'] = df_cleaned['HGB'].fillna(HGB_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'HGB'] = df_cleaned['HGB'].fillna(HGB_mean_F)
df_cleaned.loc[df_cleaned['Пол']==1, 'RBC'] = df_cleaned['RBC'].fillna(RBC_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'RBC'] = df_cleaned['RBC'].fillna(RBC_mean_F)
df_cleaned.loc[df_cleaned['Пол']==1, 'HCT'] = df_cleaned['HCT'].fillna(HCT_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'HCT'] = df_cleaned['HCT'].fillna(HCT_mean_F)
df_cleaned.loc[df_cleaned['Пол']==1, 'WBC'] = df_cleaned['WBC'].fillna(WBC_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'WBC'] = df_cleaned['WBC'].fillna(WBC_mean_F)
df_cleaned.loc[df_cleaned['Пол']==1, 'LY%'] = df_cleaned['LY%'].fillna(LY_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'LY%'] = df_cleaned['LY%'].fillna(LY_mean_F)
df_cleaned.loc[df_cleaned['Пол']==1, 'PLT'] = df_cleaned['PLT'].fillna(PLT_mean_M)
df_cleaned.loc[df_cleaned['Пол']==0, 'PLT'] = df_cleaned['PLT'].fillna(PLT_mean_F)

df_cleaned = df_cleaned.drop('NE%', axis=1)
```

### Генерация данных с использованием нормального распределения

```python
# Функция генерации синтетических данных
def generate_synthetic_data(params, size, gender_value):
    data = {}
    for param, (mean, std) in params.items():
        data[param] = np.random.normal(mean, std, size)
    data['Пол'] = [gender_value] * size
    data['Возраст'] = np.random.normal(71, 11, size)
    data['target'] = 0  # добавляем столбец target - 0 - нет лейкоза
    return pd.DataFrame(data)
```

## Обучение моделей

### Разделение данных на обучающую и тестовую выборки

При разделении необходимо учесть, чтобы данные с одним ID не попали в обе выборки.

```python
# Создаем объект GroupShuffleSplit
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

# Разделяем данные, используя идентификаторы людей
train_idx, test_idx = next(gss.split(df_shuffled, groups=df_shuffled['ID пациента']))

# Создаем тренировочные и тестовые выборки
train = df_shuffled.iloc[train_idx]
test = df_shuffled.iloc[test_idx]

print(train.shape)
print(test.shape)
```

```python
# Разделение данных обучающей выборки на признаки и целевую переменную.
X_train = train.drop(['target', 'ID пациента'], axis=1)
y_train = train['target']

# Разделение данных тестовой выборки на признаки и целевую переменную.
X_test = test.drop(['target', 'ID пациента'], axis=1)
y_test = test['target']
```

### Модель случайного леса

```python
# Создание и обучение модели случайного леса
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Прогнозирование на тестовой выборке
y_rf_pred = model.predict(X_test)

# Оценка точности модели
recall = sklearn.metrics.recall_score(y_test, y_rf_pred)
print(f'Recall: {recall:.2f}')

# Подробный отчет по классификации
print(classification_report(y_test, y_rf_pred))
```

Модель нам необходима для первичного скрининга на лейкоз с дальнейшей рекомендацией обратиться за консультацией к врачу и при необходимости дальнейшим дообследованием.
